```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('amazon_reviews.csv')
df.head()
```

```python
df.rename(columns = {'Unnamed: 0':'Id'}, inplace = True)
```

```python
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sia = SentimentIntensityAnalyzer()
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['reviewText']
    myid = row['Id']
    res[myid] = sia.polarity_scores(str(text))
```

```python
res
```

```python
print(res[0]['neg'])
```

```python
len(res)
```

```python
from decimal import Decimal

output=[]
for item in res:
  if(Decimal(res[item]['neg'])>0.5):
    output.append('Negative')
  elif(Decimal(res[item]['pos'])>0.5):
    output.append('Positive')
  elif(Decimal(res[item]['neu'])>0.5):
    output.append('Neutral')
  else:
    output.append('Neutral')
```

```python
len(output)
```

```python
df.drop(columns='Classification', inplace=True)
```

```python
df['Classification'] = output
```

```python
df.head()
```

```python
df.to_csv('/content/sentiment_analysis.csv')
```
